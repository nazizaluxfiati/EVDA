Dari data nilai di atas, dapat dilihat bahwa fitur yang digunakan untuk mengetahui apakah ada kecenderungan dosen memberikan nilai lebih tinggi atau lebih rendah adalah fitur NIDN yang kepanjangan dari Nomor Induk Dosen Nasional dengan fitur nilai. Maka langkah berikutnya hanya mengambil fitur NIDN dengan nilai, yang dihitung dengan menggunakan rata-rata nilai yang diberikan masing masing dosen ke mahasiswa dan menggunakan variansinya.

```python
rerata_nilai = df7.groupby(df7['NIDN']).mean()[['nilai']]
rerata_nilai.rename(columns={'nilai':'rerata_nilai'},inplace=True)
rerata_nilai
```

Fungsi grupby untuk mengelompokkan data dalam kumpulan item yang sama, dari atribut  nilai dengan NIDN kita bisa mencari tahu rata-rata nilai yang sering diberikan dosen ke mahasiswa. Dari data di atas dapat diketahui kalau dosen dengan NIDN 123 rata-rata nilai terbesar yang diberikan ke mahasiswa sekitar 85 dan rata-rata nilai terendah yang diberikan ke mahasiswa adalah 60 oleh dosen dengan NIDN 345.

```python
var_nilai = df7.groupby(df7['NIDN']).var()[['nilai']]
var_nilai.rename(columns={'nilai':'var_nilai'},inplace=True)
var_nilai
```

```python
summary_nilai = pd.concat([rerata_nilai, var_nilai], axis=1) #, keys=['level1','level2'])
summary_nilai
```
